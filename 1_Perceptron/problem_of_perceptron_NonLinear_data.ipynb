```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
or_data = {'ip1': [0, 0, 1, 1], 'ip2': [0, 1, 0, 1], 'op': [0, 1, 1, 1]}
or_df = pd.DataFrame(or_data)
or_df
```

```python
plt.scatter(or_data['ip1'], or_data['ip2'], c=or_data['op'])
plt.grid()
plt.show()
```

```python
and_data = {'ip1': [0, 0, 1, 1], 'ip2': [0, 1, 0, 1], 'op': [0, 0, 0, 1]}
and_df = pd.DataFrame(and_data)
and_df
```

```python
plt.scatter(and_data['ip1'], and_data['ip2'], c=and_data['op'])
plt.grid()
plt.show()
```

```python
xor_data = {'ip1': [0, 0, 1, 1], 'ip2': [0, 1, 0, 1], 'op': [0, 1, 1, 0]}
xor_df = pd.DataFrame(xor_data)
xor_df
```

```python
plt.scatter(xor_data['ip1'], xor_data['ip2'], c=xor_data['op'])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import Perceptron

clf1 = Perceptron()
clf2 = Perceptron()
clf3 = Perceptron()
```

```python
clf1.fit(or_df.iloc[:, 0:2].values, or_df.iloc[:, -1].values)
C,  A, B = clf1.intercept_, clf1.coef_[0][0], clf1.coef_[0][1]
C, A, B 
```

```python
m = -(A/B); b = -(C/B)
x_line = np.linspace(0, 1, 10)
y_line = m*x_line + b

plt.scatter(or_data['ip1'], or_data['ip2'], c=or_data['op'])
plt.plot(x_line, y_line)
plt.ylim(-0.5, 1.5)
plt.xlim(-0.5, 1.5)
plt.grid()
plt.show()
```

```python
clf2.fit(and_df.iloc[:, 0:2].values, and_df.iloc[:, -1].values)
C,  A, B = clf2.intercept_, clf2.coef_[0][0], clf2.coef_[0][1]
C, A, B 
```

```python
m = -(A/B); b = -(C/B)
x_line = np.linspace(0, 1, 10)
y_line = m*x_line + b

plt.scatter(and_data['ip1'], and_data['ip2'], c=and_data['op'])
plt.plot(x_line, y_line)
plt.ylim(-0.5, 1.5)
plt.xlim(-0.5, 1.5)
plt.grid()
plt.show()
```

```python
clf3.fit(xor_df.iloc[:, 0:2].values, xor_df.iloc[:, -1].values)
C,  A, B = clf3.intercept_, clf3.coef_[0][0], clf3.coef_[0][1]
C, A, B  # all are zero
```

```python
m = -(A/B); b = -(C/B)
x_line = np.linspace(0, 1, 10)
y_line = m*x_line + b

plt.scatter(xor_data['ip1'], xor_data['ip2'], c=xor_data['op'])
plt.plot(x_line, y_line)
plt.ylim(-0.5, 1.5)
plt.xlim(-0.5, 1.5)
plt.grid()
plt.show()
```
